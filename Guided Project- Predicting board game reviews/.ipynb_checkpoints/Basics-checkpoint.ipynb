```python
import pandas as pd

board_games = pd.read_csv("board_games.csv")

print(board_games.head(5))
```

```python
board_games = board_games.dropna(axis=0)
board_games = board_games[board_games["users_rated"] > 0]


print(board_games.head(5))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(board_games["average_rating"])
```

```python
print(board_games["average_rating"].std())
print(board_games["average_rating"].mean())
```

Makes sense, since this is a contiuous data set, and follows a somewhat normal distribution. 

We can then compare our trained model with the std to show the performance.

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=5)
cols = list(board_games.columns)
cols.remove("name")
cols.remove("id")
cols.remove("type")

numeric_cols = board_games[cols]
km.fit(numeric_cols)
labels = km.labels_
```

```python
import numpy as np
game_mean = numeric_cols.apply(np.mean, axis=1)
game_std = numeric_cols.apply(np.std, axis=1)

plt.scatter(x=game_mean, y=game_std, c=labels)
plt.show()
```
