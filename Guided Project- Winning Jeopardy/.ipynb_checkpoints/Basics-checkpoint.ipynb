```python
import pandas
import csv

jeopardy = pandas.read_csv("jeopardy.csv")

jeopardy
```

```python
jeopardy.columns = ['Show Number', 'Air Date', 'Round', 'Category', 'Value', 'Question', 'Answer']
jeopardy.columns
```

```python

import re

def normalize_text(text):
    text = text.lower()
    text = re.sub("[^A-Za-z0-9\s]", "", text)
    return text

def normalize_values(text):
    text = re.sub("[^A-Za-z0-9\s]", "", text)
    try:
        text = int(text)
    except Exception:
        text = 0
    return text
```

```python
jeopardy["clean_question"] = jeopardy["Question"].apply(normalize_text)
jeopardy["clean_answer"] = jeopardy["Answer"].apply(normalize_text)
jeopardy["clean_value"] = jeopardy["Value"].apply(normalize_values)
```

```python
def check_deducible_anwer_num(row):
    split_answer = row["clean_answer"].split(" ")
    split_question = row["clean_question"].split(" ")

    match_count = 0
    
    if "the" in split_answer:
        split_answer.remove("the")
        
    if len(split_answer) == 0:
        return 0
    
    for ans in split_answer:
        if ans in split_question:
            match_count += 1
    
    return match_count * 1.0 / len(split_answer)

answer_in_question = jeopardy.apply(check_deducible_anwer_num, axis=1)
print(answer_in_question)
print(answer_in_question.mean())
```
